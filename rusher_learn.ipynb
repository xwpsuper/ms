Python习题: 写一个简单的股票抢单器。

- 前置0. 假定已经有封装好的交易类Trader，见最后
- 任务1. 不停调用Trader抢单直到抢中为止
- 任务2. 使用多线程调用Trader, 使下单间隔为0.005秒, 抢中以后终止
- 任务3. 多阶段调用, 在开启任务的前3分钟, 下单间隔为1秒1次, 第4分钟0.2秒一次, 第5分钟0.02秒一次, 第6分钟0.002秒一次


```python
import time
import random

class Trader(object):
    def __init__(self, username):
        self.username = username
    def buy(self, symbol, price, lot):
        print(f'|{time.time() % 60:6.3f}| {self.username}买入{symbol}: {price} x {lot}')
        time.sleep(random.random() * 0.03 + 0.02)
        return True if random.random() < 0.005 else False
```

```python
t = Trader('ths')
t.buy('601001', 5.00, 1)
```

```python
while True:
    if t.buy('601001', 5.00, 1):
        break
```

```python
from concurrent.futures import ThreadPoolExecutor
pool = ThreadPoolExecutor(100)
done = False

def threadbuy(symbol, price, lot):
    global done
    done = t.buy(symbol, price, lot)
    
def add(a, b): return a + b

for i in range(100):
    pool.submit(add, 1, 1)

while not done:
    pool.submit(threadbuy, '601001', 5.00, 1)
    t0 = time.time()
    while time.time() - t0 < 0.005:
        time.sleep(0.0001)
```

### 3. 分阶段任务

```python
from concurrent.futures import ThreadPoolExecutor
pool = ThreadPoolExecutor(100)
done = False

def threadbuy(symbol, price, lot):
    global done
    done = t.buy(symbol, price, lot)

while not done:
    pool.submit(threadbuy, '601001', 5.00, 1)
    t0 = time.time()
    while time.time() - t0 < 0.005:
        time.sleep(0.00001)
```

```python
# 前3分钟, 下单间隔为1秒1次, 第4分钟0.2秒一次, 第5分钟0.02秒一次, 第6分钟0.002秒一次
intervals = [
    {'stop': 18, 'interval': 1},
    {'stop': 24, 'interval': 0.2},
    {'stop': 30, 'interval': 0.02},
    {'stop': 36, 'interval': 0.002},
]

class Rusher(object):
    def __init__(self, intervals):
        self.intervals = intervals
        self.interval = 1
        self.starttime = 0
        self.done = False
        self.trader = Trader('ths')
    
    def buy(self, symbol, price, lot):
        self.done = self.trader.buy(symbol, price, lot)
        
    def rush(self):
        self.starttime = time.time()
        while not self.done:
            passed_time = time.time() - self.starttime
            for params in self.intervals:
                if passed_time < params['stop']:
                    self.interval = params['interval']
                    break
            pool.submit(self.buy, '601001', 5.00, 1)
            time.sleep(self.interval)
```

```python
Rusher(intervals).rush()
```
